```python
import random
import time

card_deck = {
    "Diablo": {"Health": 100, "Attack": 90, "Defense": 60},
    "Medusa": {"Health": 100, "Attack": 70, "Defense": 70},
    "Jester": {"Health": 120, "Attack": 60, "Defense": 90},
    "Troll": {"Health": 150, "Attack": 40, "Defense": 94},
    "Specter": {"Health": 100, "Attack": 70, "Defense": 70},
    "Mist": {"Health": 100, "Attack": 75, "Defense": 65},
    "Savage": {"Health": 100, "Attack": 90, "Defense": 50},
    "Marauder": {"Health": 100, "Attack": 85, "Defense": 50},
    "Wimp": {"Health": 110, "Attack": 40, "Defense": 85},
    "Sorcerer": {"Health": 100, "Attack": 70, "Defense": 55}
}

deck_size = 3

def draw_cards(deck, size):
    """Draws a random sample of cards from the deck."""
    cards = random.sample(list(deck.keys()), size)
    return {card: deck[card].copy() for card in cards}

def display_card(card_name, card_details):
    """Displays the details of a card."""
    print(f"\nName: {card_name}")
    for stat, value in card_details.items():
        print(f"{stat}: {value}")

def attack_card(attacker, defender):
    """Handles the attack logic between two cards."""
    if attacker["Attack"] > defender["Health"]:
        print(f"\n{attacker_name} defeats {defender_name} with a powerful blow!")
        return True  # Defender is defeated
    else:
        damage = attacker["Attack"]
        if defender["Defense"] > 0:
            defense_reduction = min(defender["Defense"], damage)
            defender["Defense"] -= defense_reduction
            damage -= defense_reduction
            print(f"\n{attacker_name} attacks {defender_name}, reducing its defense by {defense_reduction}.")
        if damage > 0:
            defender["Health"] -= damage
            print(f"\n{attacker_name} attacks {defender_name}, dealing {damage} damage.")
        return False  # Defender is not defeated


# Game loop
while True:
    player_deck = draw_cards(card_deck, deck_size)
    opponent_deck = draw_cards(card_deck, deck_size)
    
    print("\n--- New Game ---")

    while player_deck and opponent_deck:
        # Player's turn
        print("\n--- Your Turn ---")
        print("\nYour cards:")
        for i, (card_name, card_details) in enumerate(player_deck.items()):
            print(f"{i+1}. {card_name}")
            display_card(card_name, card_details)

        opponent_card_name = random.choice(list(opponent_deck.keys()))
        opponent_card = opponent_deck[opponent_card_name]
        print("\nOpponent's card:")
        display_card(opponent_card_name, opponent_card)

        while True:
            try:
                player_choice = int(input("\nChoose a card to fight (enter the number): ")) - 1
                if 0 <= player_choice < len(player_deck):
                    break
                else:
                    print("Invalid choice. Please enter a valid number.")
            except ValueError:
                print("Invalid input. Please enter a number.")

        player_card_name = list(player_deck.keys())[player_choice]
        player_card = player_deck[player_card_name]
        
        attacker_name = player_card_name
        defender_name = opponent_card_name
        
        if attack_card(player_card, opponent_card):
            del opponent_deck[opponent_card_name]
        
        
        # Update player deck if player card's health is 0 or less
        if player_card["Health"] <= 0:
            del player_deck[player_card_name]
        
        time.sleep(3)

        # Opponent's turn (if opponent has cards left)
        if opponent_deck:
            print("\n--- Opponent's Turn ---")
            opponent_card_name = random.choice(list(opponent_deck.keys()))
            opponent_card = opponent_deck[opponent_card_name]
            print("\nOpponent's card:")
            display_card(opponent_card_name, opponent_card)
            
            print("\nYour Cards:")
            for i, (card_name, card_details) in enumerate(player_deck.items()):
                print(f"{i+1}. {card_name}")
                display_card(card_name, card_details)

            while True:
                try:
                    player_choice = int(input("\nChoose a card to fight (enter the number): ")) - 1
                    if 0 <= player_choice < len(player_deck):
                        break
                    else:
                        print("Invalid choice. Please enter a valid number.")
                except ValueError:
                    print("Invalid input. Please enter a number.")

            player_card_name = list(player_deck.keys())[player_choice]
            player_card = player_deck[player_card_name]
            
            attacker_name = opponent_card_name
            defender_name = player_card_name
            
            if attack_card(opponent_card, player_card):
                del player_deck[player_card_name]
            
            # Update opponent deck if opponent card's health is 0 or less
            if opponent_card["Health"] <= 0:
                del opponent_deck[opponent_card_name]

            print("\n--- New round will begin in 3 seconds ---")
            time.sleep(3)

    # Determine the winner
    if not opponent_deck:
        print("\n--- You Win! ---")
    else:
        print("\n--- Opponent Wins! ---")
    
    play_again = input("Play again? (y/n): ")
    if play_again.lower() != 'y':
        break
```
